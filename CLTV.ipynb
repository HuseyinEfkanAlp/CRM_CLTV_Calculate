```python
df = df_.copy()
df.head()
```

```python
df.isnull().sum() #eksik değerler
```

```python
#Invoice sütunun içinde C içeren satırları sildik çünkü C içerenler iade
#tilda  = olanları getirme
df = df[~df["Invoice"].str.contains("C", na=False)] 
df.head()
```

```python
df.describe().T
```

```python
df = df[df["Quantity"]>0]
df.describe().T
```

```python
#df deki eksik değerleri kalıcı bir şekilde sildik
df.dropna(inplace = True)
df.describe().T
```

```python
df["TotalPrice"]= df["Quantity"]*df["Price"]
df.head()
```

```python
cltv_c = df.groupby("Customer ID").agg({"Invoice": lambda x: x.nunique(),
                                        "Quantity": lambda x: x.sum(),
                                        "TotalPrice": lambda x: x.sum()})
cltv_c
```

```python
cltv_c.columns = ["total_transaction", "total_unit", "total_price"]
cltv_c
```

```python
#1Average Order Valur(total_price / total_transcation)
```

```python
cltv_c["average_order_value"] = cltv_c["total_price"]/cltv_c["total_transaction"]
```

```python
#2 Purchase Frequency (total_transaction / total_number_of_customers)
```

```python
cltv_c["purchase_frequency"] = cltv_c["total_transaction"]/cltv_c.index.nunique()
```

```python
#3 Repeat Rate (1den fazla alışveriş yapmış müşteriler / tüm müşteriler)
#Churn Rate(1- Repeat Rate)
```

```python
repeat_rate = cltv_c[cltv_c["total_transaction"]>1].shape[0]/cltv_c.shape[0]
churn_rate = 1 - repeat_rate
```

```python
#4 Profit Margin(total_price * 0.10)
```

```python
cltv_c["profit_margin"] = cltv_c["total_price"]*0.10
cltv_c
```

```python
#5 Customer Value(average_order_value * purchase_frequency)
```

```python
cltv_c["customer_value"] = cltv_c["average_order_value"]*cltv_c["purchase_frequency"]
```

```python
#6 Customer Life Time Value (cutomer_value / churn_rate) * profit_margin
```

```python

cltv_c["customer_life_time_value"] = (cltv_c["customer_value"] / churn_rate) * cltv_c["profit_margin"]
cltv_c
```

```python
cltv_c.sort_values(by = "customer_life_time_value", ascending = False).head()
```

```python
#segmentlere ayırma
cltv_c["segment"] = pd.qcut(cltv_c["customer_life_time_value"],4,labels= ["D","C","B","A"])
```
